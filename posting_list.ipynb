```python
post_list = {}

for index, row in df.iterrows():
    cod_modelo_color = row['cod_modelo_color']

    outs_appear = set(outfit[outfit['cod_modelo_color'] == cod_modelo_color]['cod_outfit'].unique())

    # Inicializar el conjunto sin incluir el cod_modelo_color actual
    post_list[cod_modelo_color] = set()

    for num in outs_appear:
        products_in_outfit = set(outfit[outfit['cod_outfit'] == num]['cod_modelo_color'])
        products_in_outfit.discard(cod_modelo_color)
        post_list[cod_modelo_color].update(products_in_outfit)
```
